```python
from sklearn import svm
import pandas as pd
import numpy as np

C = 1000
gamma = .00001
model = svm.SVC(C=C, gamma=gamma, kernel='rbf', probability=True)
```

```python
train = pd.read_csv('/content/drive/MyDrive/DATA/diabetes_train.csv')

test = train.iloc[(len(train)-10):(len(train)), ] #takes last 10 entries in training set 'train'
train = train.iloc[0:(len(train)-10), ] #takes all entries up to before last 10 in 'test' set
test #check if last 10 entries are in testing set 'test'
```

```python
model = model.fit(train.iloc[:, 0:8], train.iloc[:, 8]) #train model for data 0 thru 7, based on output in column 8
pred = model.predict(test.iloc[:, 0:8]) #make prediction for testing set 'test'
pred
```

```python
from sklearn.metrics import confusion_matrix
true = test.iloc[:,8] 

tn, fp, fn, tp = confusion_matrix(true, pred).ravel() #test predictions against 'true' and print results

print("True Neg.  = ", tn," False Pos. = ", fp)
print("False Neg. = ", fn," True Pos.  = ", tp)
print("\n")
print("Prediction Success = ", (tp+tn), "/", (tp+tn+fp+fn))
```
